```python
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable
from matplotlib.colors import Normalize
from matplotlib.colors import LinearSegmentedColormap
```

```python
image_name = "24"

truth_path = f"inference/truth/{image_name}.png"
attention_path = f"inference/attention/latest/{image_name}.png"
mean_path = f"inference/mean/latest/{image_name}.png"
max_path = f"inference/max/latest/{image_name}.png"

truth_image = Image.open(truth_path)
attention_image = Image.open(attention_path)
mean_image = Image.open(mean_path)
max_image = Image.open(max_path)
```

```python
cmap_custom = LinearSegmentedColormap.from_list(
    'custom',
    [(int(255 * score), 0, int(255 * (1 - score))) for score in range(0, 1.1, 0.1)]
)
```

```python
fig, axs = plt.subplots(1, 4)

norm = Normalize(vmin=0, vmax=1)
sm = ScalarMappable(norm=norm, cmap=cmap_custom)
sm.set_array([])

fig.colorbar(sm, ax=axs)

for ax in axs:
    ax.set_axis_off()
axs[0].set_title("truth")
axs[0].imshow(truth_image)
axs[1].set_title("attention")
axs[1].imshow(attention_image)
axs[2].set_title("mean")
axs[2].imshow(mean_image)
axs[3].set_title("max")
axs[3].imshow(max_image)
```
